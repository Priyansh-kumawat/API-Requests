```python
URL = 'https://api.mfapi.in/mf/102907'
response = requests.get(URL)
response.json()
```

Explanation:

Reliance Mutual Fund: The asset management company managing the mutual fund.

Open Ended Schemes: These are mutual fund schemes that do not have a fixed maturity date and can be bought and sold at any time.

Income Category: This category typically includes funds that aim to provide regular income to investors, usually through dividends or interest.

Scheme Code (102907): A unique identifier for the specific mutual fund scheme.

Scheme Name: Specifies the exact mutual fund plan, which in this case tracks the Sensex index and has a dividend option.

Date (17-04-2008): The specific date for which the NAV is provided.

NAV (23.46360): The Net Asset Value of the mutual fund on the given date, indicating the per-unit price of the fund.

The NAV value helps investors understand the performance and current value of the mutual fund on the specified date.

```python
data = response.json()["data"]
data
```

```python
data = response.json()["data"]

temp_dict = {}
for entry in data:
    temp_dict[entry["date"]] = entry["nav"]

temp_dict
```

# the loop iterates over each entry in the data list 
# and it accesses the "date" and "nav" keys of each dictionary to populate temp_dict.

```python
temp_df = pd.DataFrame({
                        "Date" : temp_dict.keys(),
                        "Net Asset Value" : temp_dict.values()})
temp_df["Date"]=pd.to_datetime(temp_df["Date"], format = '%d-%m-%Y')
temp_df["Net Asset Value"]=temp_df["Net Asset Value"].astype(float)
temp_df
```

```python
plt.figure(figsize=(10,4))
plt.plot(temp_df["Date"],temp_df["Net Asset Value"],marker=0)
plt.title= "NAV over Time"
plt.xlabel= "date"
plt.ylabel = "NAV"
plt.grid(True)
plt.show()
```

# now we can save it as csv file

```python
temp_df.to_csv('nav_data.csv',index=False)
```
